# Vertical Line Test

## 1.1 Create two graphs, one that passes the vertical line test and one that does not.

```
import matplotlib.pyplot as plt
import numpy as np

x = np.array(range(5))
y = [2*i for i in x]

plt.xlim([0,6])
plt.ylim([0,10])
plt.plot(x,y)
plt.grid()
plt.title('Passes Vertical Line Test')
plt.show()
```

```
theta = np.linspace(0, 2*np.pi, 100)

r = np.sqrt(1.0)

x1 = r*np.cos(theta)
x2 = r*np.sin(theta)

plt.xlim([-2,2])
plt.ylim([-2,2])
plt.plot(x1,x2)
plt.grid()
plt.title('Does Not Pass Vertical Line Test')
plt.show()
```

## 1.2 Why are graphs that don't pass the vertical line test not considered "functions?"

A function is a relation between a set of inputs that are mapped to exactly one output therefore, multiple instances of crossing on a vertical line test would mean there are more than 1 output for an input.

# Functions as Relations

## 2.1 Which of the following relations are functions? Why?

\begin{align}
\text{Relation 1: } \{(1, 2), (3, 2), (1, 3)\}
\\
\text{Relation 2: } \{(1, 3), (2, 3), (6, 7)\}
\\
\text{Relation 3: } \{(9, 4), (2, 1), (9, 6)\}
\\
\text{Relation 4: } \{(6, 2), (8, 3), (6, 4)\}
\\
\text{Relation 5: } \{(2, 6), (2, 7), (2, 4)\}
\end{align}

Relation 1 IS NOT a function because the first and third instance containing x = 1 map to different outputs so they will not pass a vertical line test.

Relation 2 IS a function because the x values map to one output - the first two instances where both ordered pairs have the same output does not negate it being a function.

Relation 3 IS NOT a function because the first and third input '9' result in different outputs.

Realtion 4 IS NOT a function because the input '6' results in two values '2' and '4'.

Realtion 5 IS NOT a function because each input is the same yet, results in a different output in all three instances.

\begin{align}

First Function:

Second Function:

$q(\begin{bmatrix}x_1\\x_2\\x_3\end{bmatrix})=x_1\begin{bmatrix}1\\0\\0\end{bmatrix}+x_2\begin{bmatrix}0\\1\\0\end{bmatrix}+x_3\begin{bmatrix}0\\0\\1\end{bmatrix}
\\
q(\begin{bmatrix}1\\0\\0\end{bmatrix})=\begin{bmatrix}4(1)+(0)+2(0)\\2(0)-(1)+3(0)\\5(1)-2(0)+(0)\end{bmatrix}=\begin{bmatrix}4\\-1\\5\end{bmatrix}

q(\begin{bmatrix}0\\1\\0\end{bmatrix})=\begin{bmatrix}4(0)+(1)+2(0)\\2(1)-(0)+3(0)\\5(0)-2(0)+(1)\end{bmatrix}=\begin{bmatrix}1\\2\\1\end{bmatrix}

\\
\rightarrow T=\begin{bmatrix}4&1&2\\-1&2&3\\5&1&-2\end{bmatrix}

```
import os.path
from urllib.request import urlopen
from zipfile import ZipFile
from io import BytesIO
import pandas as pd

# Read National Data 
national_url = 'https://www.huduser.gov/portal/datasets/hads/hads2013n_ASCII.zip'
national_file = 'thads2013n.txt'

if os.path.exists(national_file):
    national = pd.read_csv(national_file)
else: 
    z_national = urlopen(national_url)
    zip_national = ZipFile(BytesIO(z_national.read())).extract(national_file)
    national = pd.read_csv(zip_national)
```

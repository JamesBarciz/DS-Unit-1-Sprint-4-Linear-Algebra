## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
h = np.matrix('12 11; 7 10')
j = np.matrix('0 1 2; 7 10 4; 3 2 0')

det_h = np.linalg.det(h)
det_j = np.linalg.det(j)

print('|H| =', det_h)
print('|J| =', det_j)
```

```
print('H^-1 =\n', np.linalg.inv(h))
print('\nJ^-1 =\n', np.linalg.inv(j))
```
